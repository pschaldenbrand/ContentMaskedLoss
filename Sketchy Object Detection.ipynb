```python
from sketchy.classifier import SketchyClassifier
from paint import *

import pickle
```

```python
classifier = SketchyClassifier()

renderer_fn = 'renderer_constrained.pkl'

actor_fns = [
#              'model/Paint-run48/actor.pkl',
#              'model/Paint-run49/actor.pkl',
#              'model/Paint-run50/actor.pkl',
#              'model/Paint-run51/actor.pkl',
#              'model/Paint-run52/actor.pkl'
                 'pretrained_models/gan/actor.pkl',
                 'pretrained_models/l2/actor.pkl',
                 'pretrained_models/l1/actor.pkl',
                 'pretrained_models/cm/actor.pkl',
                 'pretrained_models/cml1/actor.pkl'
            ]
labels = ['GAN', 'L2', 'L1*', 'CM', 'CM + L1*']
actor_results = [[] for a in actor_fns]

# Get the files for the test
test_file_names = []
test_file_classes = {} # {file_name : class_name}

for class_name in SketchyClassifier.class_names:
    class_dir = os.path.join(SketchyClassifier.sketchy_img_dir, class_name)

    # Sort filenames for consistency in train/val split
    class_img_ind = 0
    for img_fn in sorted(os.listdir(class_dir)):
        if class_img_ind >= 90:
            img_path = os.path.join(class_dir, img_fn)
            test_file_names.append(img_path)
            test_file_classes[img_path] = class_name
        class_img_ind += 1

i = 0
for fn in test_file_names:
    img_class = test_file_classes[fn]
    
    for actor_ind in range(len(actor_fns)):
        actor_fn = actor_fns[actor_ind]
        
        canvas_ind, last_canvas = paint_until_object_detected(fn, actor_fn, renderer_fn, img_class, classifier, \
                                                              max_big_strokes=750, white_canvas=True, \
                                                              div=3, max_small_strokes=750 )
        print('i ', i, '\t#', canvas_ind, '\t', img_class, sep='')
        
        result = {}
        result['canvas_ind'] = canvas_ind
        result['last_canvas'] = last_canvas
        result['file_name'] = fn
        result['class_name'] = img_class
        result['actor_fn'] = actor_fn

        actor_results[actor_ind].append(result)
    i += 1
    if (i % 10) == 0:
        pickle.dump( actor_results, open( "object_detection_results_sketchy_ca.p", "wb" ) )

pickle.dump( actor_results, open( "object_detection_results_sketchy_ca.p", "wb" ) )
```

```python
c_norm = cv2.resize(last_canvas, (classifier.width, classifier.width))
c_norm = classifier.normalize(c_norm)
pred_class = classifier.classify(c_norm.unsqueeze(0).to(device))
print(img_class, pred_class)
pred = classifier(c_norm.unsqueeze(0).to(device))
# pred = torch.exp(pred)
# pred = -1 * torch.log(pred / torch.sum(pred))
# pred = pred / torch.sum(pred)
print(pred, torch.sum(pred), classifier.class_names[torch.argmax(pred)], img_class, sep='\n')
```

```python
# Are there categories that are particularly difficult to classify? TEST IMAGES
class_correct = {}
class_incorrect = {}

for fn in test_file_names:
    img_class = test_file_classes[fn]
    
    if img_class not in class_correct.keys():
        class_correct[img_class] = 0
    if img_class not in class_incorrect.keys():
        class_incorrect[img_class] = 0
        
    img = cv2.imread(fn, cv2.IMREAD_UNCHANGED)
    img = img[:,:,::-1] # BGR to RGB
    img = cv2.resize(img, (SketchyClassifier.width, SketchyClassifier.width))
    img = SketchyClassifier.normalize(img)
    prediction, confidence = classifier.classify(img.unsqueeze(0).to('cuda'))
    if prediction[0] == img_class:
        class_correct[img_class] += 1
    else:
        class_incorrect[img_class] += 1
print(class_correct)
```

```python
for c in sorted(class_correct, key=class_correct.get, reverse=True):
    print(c, class_correct[c])
```

```python
# Are there categories that are particularly difficult to classify? ALL IMAGES
class_correct_all = {}
class_incorrect_all = {}

for class_name in SketchyClassifier.class_names:
    class_dir = os.path.join(SketchyClassifier.sketchy_img_dir, class_name)

    for img_fn in sorted(os.listdir(class_dir)):
        img_path = os.path.join(class_dir, img_fn)
    
        if class_name not in class_correct_all.keys():
            class_correct_all[class_name] = 0
        if class_name not in class_incorrect_all.keys():
            class_incorrect_all[class_name] = 0

        img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        img = img[:,:,::-1] # BGR to RGB
        img = cv2.resize(img, (SketchyClassifier.width, SketchyClassifier.width))
        img = SketchyClassifier.normalize(img)
        prediction, confidence = classifier.classify(img.unsqueeze(0).to('cuda'))
        if prediction[0] == class_name:
            class_correct_all[class_name] += 1
        else:
            class_incorrect_all[class_name] += 1

for c in sorted(class_correct_all, key=class_correct_all.get, reverse=True):
    print(c, class_correct_all[c])
```
